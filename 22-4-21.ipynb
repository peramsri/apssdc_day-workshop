# Bias
-Bias is the simplifying assumption made by the model to make the target function easier to approximate

# Variance
-Variance is the amount that the estimate of the target fuction will change given different training data

## Overfitting
-Good performance on the training data.low bias and high variance will get the occurance of overfitting


 ## Underfitting
-poor performance on the training data and poor generalization to other data. 
-High bias and low variance

## KNN->K-Nearest Neighbour Classifier
- KNN is easy to implement
- Support Classification and Regression
- knn algorithm assumes that the similar things exists nearer to each other
### How does knn works
1. select k number of neighbours,i.e k=3,5
2. calculate the euclidien distance b/w data
3. assign the new data point to the majority nearest neighbours class


```python
df=pd.read_csv("https://raw.githubusercontent.com/nagamounika5/Datasets/master/Tshirt_size.csv")
df
```
